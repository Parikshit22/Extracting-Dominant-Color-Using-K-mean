```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('download.jpg')
original_shape = img.shape
print(original_shape)
im = img.reshape(-1,3)
print(im.shape)
```

```python
k = 10
from sklearn.cluster import KMeans
km = KMeans(n_clusters=k)
```

```python
km.fit(im)
```

```python
centers = km.cluster_centers_
```

```python
cen = np.array(centers,dtype = 'uint8')
```

```python
print(cen)
```

```python
ix = 1
color = []
for i in cen:
    plt.subplot(1,k,ix)
    plt.axis('off')
    color.append(i)
    ix +=1
    ze = np.zeros((100,100,3),dtype = 'uint8')
    ze[:,:,:] = i
    plt.imshow(ze)
```

```python
new_img = np.zeros((50176,3),dtype = 'uint8')
```

```python
label = km.labels_
```

```python
for i in range(new_img.shape[0]):
    new_img[i] = color[label[i]]
new_img = new_img.reshape(original_shape)
plt.imshow(new_img)
```
